```python
import cv2

video_path = r"C:\Users\Victus\OneDrive\Desktop\ML Project\Accident Detection from CCTV\videos\accident1.mp4.mp4"
cap = cv2.VideoCapture(video_path)

ret, prev_frame = cap.read()
if not ret:
    print("‚ùå Cannot read video")
    exit()

prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Frame difference
    diff = cv2.absdiff(prev_gray, gray)

    # Thresholding (important step)
    _, thresh = cv2.threshold(diff, 25, 255, cv2.THRESH_BINARY)

    # Remove noise using morphology
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
    clean = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

    cv2.imshow("Original Video", frame)
    cv2.imshow("Clean Motion Mask", clean)

    prev_gray = gray

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
